```python
# Changing date column from object to datetime format
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')
```

```python
# Upon first look at the data some columns has '$' and ','. 
df[['unit_cost', 'unit_price', 'profit', 'cost', 'revenue']].head(5)
```

```python
# Stripping $ and commas from the columns 'unit_cost', 'unit_price', 'profit', 'cost' and, 'revenue'. 
df['unit_cost'] = df['unit_cost'].str.lstrip('$')
df['unit_cost'] = df['unit_cost'].str.replace(',','')
df['unit_cost'] = df['unit_cost'].astype(float)
```

```python
df['unit_price'] = df['unit_price'].str.lstrip('$')
df['unit_price'] = df['unit_price'].str.replace(',', '')
df['unit_price'] = df['unit_price'].astype(float)
```

```python
df['profit'] = df['profit'].str.lstrip('$')
df['profit'] = df['profit'].str.replace(',', '')
df['profit'] = df['profit'].astype(float)
```

```python
df['cost'] = df['cost'].str.lstrip('$')
df['cost'] = df['cost'].str.replace(',', '')
df['cost'] = df['cost'].astype(float)
```

```python
df['revenue'] = df['revenue'].str.lstrip('$')
df['revenue'] = df['revenue'].str.replace(',', '')
df['revenue'] = df['revenue'].astype(float)
```

```python
# The column 'product_description' contains information on model, size, and color.
# Splitting the column to make new columns containing information on model, size, and color.
df[['model', 'bike_size']] = df['product_description'].str.split(',', expand = True)
df[['bike_model', 'bike_color']] = df['model'].str.split(' ', expand = True)
```

```python
# Checking the unique values in the 'customer_age' column
df['customer_age'].unique()
```

```python
# Defining age groups and then creating and populating 'age_group' column 
def categorize_age(age):
    if age < 25:
        return 'Youth'
    elif age <= 34:
        return 'Young Adults'
    else:
        return 'Adults'

df['age_group'] = df['customer_age'].apply(categorize_age)
```
