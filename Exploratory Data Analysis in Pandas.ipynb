```python
# Looking at the number of Countries in each Continent
continent = df.groupby('Continent').size()
sorted_result = continent.sort_values(ascending=False)
sorted_result
```

```python
# Grouping by continent, and sorting oby '2022 Population'
df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
# For looking at the percentage of population of the whole poplulation in each continent
#per_continent = df.groupby('Continent')['World Population Percentage'].sum()
#sort_per_continent = per_continent.sort_values(ascending = False)
#sort_per_continent
```

```python
# For looking at the percentange share of population and Density (per km²) only
per_continent = df.groupby('Continent').agg({
    'World Population Percentage': 'sum',
    'Density (per km²)': 'mean',
    'Growth Rate': 'mean'
})

sort_per_continent = per_continent.sort_values(by='World Population Percentage', ascending=False)
sort_per_continent
```

    The output shows that Asia has the largest share of world population, while Oceania has the lowest.
    The average population density of each country is far more greater than that of the rest of the continents.
    The average population growth across the continents is almost the same, though Africa is leading the pack by a tiny margin.
